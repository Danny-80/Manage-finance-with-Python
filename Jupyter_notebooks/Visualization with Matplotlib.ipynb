### Customization

```python
titanic.age.plot( figsize=(12,8), fontsize=13, c='green',linestyle= '-', xlim= (0,900), ylim = (0,89))  
plt.title('Titanic - Ages', fontsize=12)
plt.legend(loc=2, fontsize=8)
plt.xlabel('Passenger No', fontsize=10)
plt.ylabel('Passenger Age', fontsize=10)
plt.grid()
plt.show()
```

```python
xticks = [x for x in range(0,900, 50)]
xticks
```

```python
yticks = np.arange(0,81,5)
yticks
```

```python
titanic.age.plot( figsize=(12,8), fontsize=13, c='green',linestyle= '-', xlim= (0,900), ylim = (0,89), xticks=xticks, yticks=yticks)  
plt.title('Titanic - Ages', fontsize=12)
plt.legend(loc=2, fontsize=8)
plt.xlabel('Passenger No', fontsize=10)
plt.ylabel('Passenger Age', fontsize=10)
plt.grid()
plt.show()
```

```python
titanic.age.plot( figsize=(12,8), fontsize=13, c='green',linestyle= '-', xlim= (0,900), ylim = (0,89),
                 xticks=xticks, yticks=yticks, rot='45')  
plt.title('Titanic - Ages', fontsize=12)
plt.legend(loc=2, fontsize=8)
plt.xlabel('Passenger No', fontsize=10)
plt.ylabel('Passenger Age', fontsize=10)
# plt.grid()
plt.show()
```

### Categorical Plots

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.stripplot(data=titanic, x='sex', y='age',jitter=True, hue='survived', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.swarmplot(data=titanic, x='sex', y='age', hue='pclass', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data=titanic, x='sex', y='age', hue='pclass', dodge=True)
sns.swarmplot(data=titanic, x='sex', y='age', hue='pclass', dodge=True, c='yellow')

plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data=titanic, y='age', x='pclass', hue='sex', dodge=True, split= True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.barplot(data=titanic, x='pclass', y='age', hue='sex', dodge=True, orient='v')

plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.pointplot(data=titanic, x='pclass', y='age', hue='sex', dodge=True)

plt.show()
```
